# Transformada Z

## Conceito

Uma transformada “transforma o dominio do sinal”; Temos o mesmo sinal so que escrito em outro domínio.

A transformada $z$ pega um sinal de tempo discreto ($N$) e escreve esse sinal em um dominio $z$ (uma variável que é um número complexo)

$z=|z|e^{j\phi}$

## Aplicação

Dado um sinal $x(n)$, a sua transformada $X(z)=\sum_{n=-\infty}^{+\infty}x[n]*z^{-n}$

É importante lembrar que o dominio z não é infinito, então a transformada z resulta em um sinal que cuida das regiões de convergência do sinal.

a faixa de valores de z que fazem a resposta do sinal $X(z)$ não explodirem pro infinito é chamada de **região de convergência**.

$R_{x-}<|z|<R_{x+}$, sendo os $R_x$ números não negativos

### Polos e zeros

Com a região de convergencia podemos encontrar os:

- Zeros: são valores de entrada que fazem o ******************numerador****************** da fração resultante da transformada z ser zero
- Polos: são valores de entrada que fazem o **denominador** da fração resultante da transformada z ser zero

É importante lembrar que para encontrar a regiao de convergencia precisamos ver o **************************modulo do número complexo**************************, que pode ser representado por uma seta no gráfico do plano complexo (onde o eixo x é a parte real e o eixo y é a parte imaginária)

## Propriedades importantes da transformada z

É dito no livro que as propriedades fundamentais da transformada z são derivadas das propriedades fundamentais das séries de Fourier discretas estudadas em sinais e sistemas.

Uma propriedade muito interessante é a propriedade 8 chamada ********************covolução********************. Ela nos diz que, a transformada de uma **covolução de sinais** $x_1$  e $x_2$ é uma multiplicação das transformadas de cada sinal separado:

$Z[x_1(n)*x_2(n)]=X_1(z)X_2(z)$

Isso ajuda bastante para o cálculo de algumas transformadas.

# Transformada Z Inversa

A transformada Z inversa é um procedimento computacional muito complexo, que requer a resolução de integrais complexas.

A ideia central sobre a transformada z inversa é a seguinte:

Quando $X(z)$ é uma função racional do domínio $z^{-1}$, ela pode ser descrita como uma soma de simples fatores:

$Z\{x[n]\}=X[z]=\sum^\infty_{n=-\infty}x[n]z^{-n}$

Lembrando que z é um número complexo:

$z=e^{j\Omega}$

# Representação de sistemas no domínio Z

Como já sabemos, sistemas são entidades que ********************processam******************** sinais de entrada para produzir um sinal de saída. Vimos em sinais e sistemas várias propriedades de sistemas (podem ser contínuos, variantes, invariantes, etc.).

Então, com o domínio Z não sera diferente. Da mesma forma que podemos analisar sinais no dominio Z, podemos também analisar sistemas no domínio Z (com algumas peculiaridades).

A primeira é que para calcular a saída de um sistema dado um sinal de entrada precisamos fazer algumas mudanças.

Normalmente, calculamos a saída $y[z]$ de um sistema LIT (linear e invariante do tempo) dessa forma:

$x[n]*h[n]=y[n]$ (onde $h[n]$ é a resposta ao impulso do sistema e $*$ representa a ********************covolução******************** de dois sinais)

No domínio z, a covolução de sinais é representada pela simples multiplicação de sinais. Portanto temos que:

$X(z) H(z)=Y(z)$

$H(z)=\frac{Y(z)}{X(z)}$

### Representação da equação de diferença do sistema

Quando o sistema LIT analisado é representado por uma equação diferencial, a resposta ao impulso $H(z)$ pode ser simplesmente calculada passando a transformada z nos dois lados da equação.

### Representação de função de transferência

A função de transferência é representada por $H(z)=\frac{Y(z)}{X(z)}$

Se a ROC de $H(z)$ incluir um círculo unitário ($z =e^{j\omega}$), então podemos extrair uma função de resposta em frequência da função de transferência: $H(e^{j\omega})$, que é composta por três partes:

$H(e^{j\omega})=[0\ or\ \pi]+[(N-M)\omega]+\sum_1^M (e^{j\omega}-z_k)-\sum_1^M(e^{j\omega}-p_k)$

Sendo a primeira constante, a segunda linear e a terceira não linear.

### Estabilidade e causalidade

Para sistemas lineares e invariantes no tempo a estabilidade BIBO (bounded-input, bounded-output) só pode existir se o círculo representado por $|z|=1$ se encontrar dentro do ROC de $H(z)$

Isso é chamado de “estabilidade no domínio z” e é formada pelos dois teoremas:

- Um sistema LIT é estável se e somente se o círculo unitário estiver contido no ROC de $H(z)$
- Um sistema causal LIT é estável se e somente se a função $H(z)$ tiver ************todos************ os polos dentro docírculo unitário ($|z|=1$)

# Capitulo 5

A transformada discreta de Fourier nos dá a representação no domínio da frequência para sequências.

A transformada de Fourier é usada para representar uma função geral (não necessariamente periódica) por uma superposição contínua ou integral de exponenciais complexos. Portanto a série pode ser representada no domínio das frequências.

Nas séries de Fourier, sabemos que funções periódicas podem ser sintetizadas como combinações lineares de exponenciais complexas.

A introdução do capitulo nos explica como podemos sintetizar a transformada discreta de Fourier para ela ser resolvida por um programa de computador, pois na definição formal ela consiste em um somatório infinito de funções exponenciais complexas.

Essa “simplificação” é chamada de **serie discreta de Fourier (DFS).**

Quando essa técnica é usada para sequências de duração finita, usamos uma nova transformada chamada **********transformada discreta de fourier (DFT).**********

Porém, mesmo com essas abstrações a computação desses métodos é bem pesada, e pra isso foram criados algoritmos que visam melhorar a performance dessas operações.

Na DFT, cada saída numérica discreta condiz a uma certa frequência. A amplitude de uma frequência da onda é calculada usando o valor absoluto do número, e a fase é obtida calculando o ângulo desse número.Tendo como base a série anterior e aplicando a função Transformada Rápida de Fourier.

## DFT

A descrição matemática da DFT é a seguinte:

$X(k)=\sum_{n=0}^{N-1}x(n)e^{-j\frac {2\pi}Nnk}$

## Exemplo do DFT

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
def DFT(x):
    """
    Function to calculate the
    discrete Fourier Transform
    of a 1D real-valued signal x
    """

    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)

    X = np.dot(e, x)

    return X
```

```python
##Sinal de exemplo

sr = 100
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 4
x += np.sin(2*np.pi*freq*t)

freq = 7
x += 0.5* np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
## Geracao do espectro de frequencia do sinal x[n]

X = DFT(x)

N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.ylabel("Amplitude")
plt.xlabel("Frequência (Hz)")
plt.show()
```

## Exemplo de DFS com a biblioteca do numpy

```python
fft = np.fft.fft(x)
T = t[1] - t[0]
N = x.size
f = np.fft.fftfreq(len(x), T)
frequencias = f[:N // 2]
amplitudes = np.abs(fft)[:N // 2] * 1 / N
plt.ylabel("Amplitude")
plt.xlabel("Frequência (Hz)")
plt.bar(frequencias, amplitudes, width=1.5)
```

# Implementação de filtros discretos no tempo

Nesse capitulo, iremos fazer efetivamente o processamento de sinais digitais. Os componentes que processam esses sinais são chamados ****************filtros.****************

O processamento digital de sinais vem com vários gargalos da computação finita, onde precisamos fazer aproximações e trabalhar com uma precisão aritmética finita.

## Estrutura dos filtros

Os filtros são basicamente sistemas LIT. Eles são compostos pelas seguintes funções:

- Somadora: recebe duas entradas e uma saída. (Para filtros com mais de duas entradas, os somadores são concatenados).
- Multiplicadores (ganho): Recebe um sinal de entrada e adiciona um ganho nesse sinal.
- Delay: Recebe uma entrada e adiciona um delay no seu tempo (positivo ou negativo). Basicamente atrasa ou adianta o sinal.

### Filtro IIR

A função de um filtro IIR (Impulse Response Infinito) é dado por:

$H(z)=\frac{B(z)}{A(z)}=\frac{b_0+b_1z^{-1}+...+b_Mz^{-M}}{1+a_1z^{-1}+...+a_Nz^{-N}}$, onde $a_0=1$

$b_n$ e $a_n$ são os coeficientes do filtro.

Os filtros IIRs podem ser estruturados de três diferentes formas:

- Forma direta
- Em cascata
    - Nesta forma, a função do sistema H(z) na equação é fatorada em seções menores de 2ª ordem, chamadas “biquads”
- Forma paralela
    - é semelhante à forma em cascata, mas após fatorização, uma expansão de fração parcial é usada para representar H(z) como uma soma de seções menores de 2ª ordem.

Uma informação interessante é que as funções IIR são muito usadas no mundo musical, para emular o comportamento de amplificadores e outros componentes.

**No processamento digital de sinal a resposta ao impulso, ou função de resposta ao impulso (IRF), de um sistema dinâmico é sua saída quando apresentada a um breve sinal de entrada, chamado de impulso (δ(t)). Mais geralmente, uma resposta ao impulso é a reação de qualquer sistema dinâmico em resposta a alguma mudança externa.**

### Filtro FIR

A função de um filtro FIR se parece com o seguinte:

$H(z)=\sum_{n=0}^{M-1}b_nz^{-n}$

Assim como o IIR, eles podem ser estruturados da seguinte forma:

- Forma direta
- Em cascada
- Forma linear
- Form “frequency-sampling”

## Exemplo de filtro IIR

```python
np.random.seed(42)
fs = 30  # Frequencia de 30hz
ts = np.arange(0, 5, 1.0 / fs)  # time vector - 5 seconds
ys = np.sin(2*np.pi * 1.0 * ts)  # sinal sem ruido
yerr = 0.5 * np.random.normal(size=len(ts))  # sinal com ruido
yraw = ys + yerr
```

```python
plt.figure(figsize = (8, 6))
plt.plot(ts, ys, 'r')
plt.ylabel('Sinal sem ruido')

plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.plot(ts, yraw, 'r')
plt.ylabel('Sinal com ruido')

plt.show()
```

```python
## Agora iremos passar o filtro no sinal com ruido
b, a = scipy.signal.iirfilter(4, Wn=2.5, fs=fs, btype="low", ftype="butter")
print(b, a, sep="\n")
y_lfilter = scipy.signal.lfilter(b, a, yraw)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(ts, yraw, 'r', label = 'Sinal com ruido')
plt.plot(ts, y_lfilter, alpha=0.8, lw=3, label="Usando o filtro da scipy")
plt.legend(loc="lower center", bbox_to_anchor=[0.5, 1],
           ncol=2, fontsize="smaller")
plt.show()
```
